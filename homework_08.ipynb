```python
from ucimlrepo import fetch_ucirepo 
import numpy as np 
import pandas as pd
```

```python
# fetch dataset 
credit_approval = fetch_ucirepo(id=27) 
```

```python
# data (as pandas dataframes) 
train = credit_approval.data.features 
sample = credit_approval.data.targets 
```

```python
mylist = []

for col in train.select_dtypes('object').columns:
    if len(train[col].value_counts()) == 2 or len(train[col].value_counts()) == 3:
        mylist.append(col)

len(mylist)
```

```python
mylist
```

```python
OWN_HOME = train['A12'].tolist()

FOREIGN_WORKER = train['A10'].tolist()

FAMILY_STATUS = train['A9'].tolist()

STUDENT  = train['A5'].tolist()

CAR_AVAIL  = train['A4'].tolist()

GENDER  = train['A1'].tolist()

train_own_home = [1 if x == 'f' else 0 for x in OWN_HOME]

train_foreign_worker = [1 if x == 'f' else 0 for x in FOREIGN_WORKER]

train_family_status = [1 if x == 'f' else 0 for x in FAMILY_STATUS]

train_student = [1 if x == 'p' or x == 'gg' else 0 for x in STUDENT ]

train_car_avail = [1 if x == 'u' else 0 for x in CAR_AVAIL]

train_gender = [1 if x == 'b' else 0 for x in GENDER]

train['A12'] = np.asarray(train_own_home)

train['A10'] = np.asarray(train_foreign_worker)

train['A9'] = np.asarray(train_family_status)

train['A5'] = np.asarray(train_student)

train['A4'] = np.asarray(train_car_avail)

train['A1'] = np.asarray(train_gender)
```

```python
train
```

```python
train.drop(['A6', 'A7', 'A13'], axis= 1, inplace=True)
```

```python
train = train.rename(
    columns={
        "A15": "DAYS_OVERDUE_NUMBER",
        "A14": "TOTAL_LOAN_SUM",
        "A12": "OWN_HOME",
        "A11": "CURR_BANK_CALLS",
        "A10": "FOREIGN_WORKER",
        "A9": "FAMILY_STATUS",
        "A8": "PREVIOUS_BANK_RATE",
        "A5": "STUDENT",
        "A4": "CAR_AVAIL",
        "A3": "CURRENT_BANK_RATE",
        "A2": "CURRENT_MONTH_LOAN",
        "A1": "MALE_GENDER",
    }
)
```

```python
sample['A16'].value_counts()
```

```python
target_value  = sample['A16'].tolist()

target_sample = [1 if x == '+' else 0 for x in target_value]

sample['A16'] = np.asarray(target_sample)

sample = sample.rename(columns={'A16': 'TARGET'})

sample.head()
```

```python
X = train
y = sample['TARGET']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_auc_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
model = DecisionTreeClassifier(criterion='entropy', max_depth=10)
```

```python
model.fit(X_train, y_train)
```

```python
preds = model.predict_proba(X_test)[:,1]
roc_auc_score(y_test, preds)
```

```python
feature_importances = pd.Series(model.feature_importances_, index=X_test.columns)
```

```python
feature_importances.sort_values()[-10:].plot.bar()
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
clf = DecisionTreeClassifier().fit(X, y)

plt.figure(figsize=(45,20))
plot_tree(clf, filled=True, fontsize=6)
plt.savefig('tree_high_dpi', dpi=150)
```

```python
from DecisionTree import DecisionTree
```

```python
no_skl = DecisionTree(max_depth = 5)
```

```python
skl = DecisionTreeClassifier(max_depth = 5)
```

```python
X, y = train_upd.values, sample['TARGET'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)
```

```python
no_skl.fit(X_train, y_train)
```

```python
skl.fit(X_train, y_train)
```

```python
predictions_no_skl = no_skl.predict(X_test)
```

```python
predictions_skl = skl.predict(X_test)
```

```python
def accuracy(y_test, y_pred):
    return np.sum(y_test == y_pred) / len(y_test)
```

```python
acc_no_skl = accuracy(y_test, predictions_no_skl)
```

```python
acc_skl = accuracy(y_test, predictions_skl)
```

```python
print(f"Accuracy without using sklearn: {acc_no_skl}")
print(f"Accuracy with using sklearn: {acc_skl}")
```
